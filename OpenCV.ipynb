```python
from PIL import Image
import matplotlib.pyplot as plt
import cv2
import numpy as np

resim=np.zeros((300,300,3))
plt.imshow(resim)
```

```python
cv2.line(resim,(0,0),(100,100),(255,0,0),3)
plt.imshow(resim)
```

```python
cv2.rectangle(resim,(200,200),(300,250),(0,255,0),3)
plt.imshow(resim)
```

```python
cv2.circle(resim,(100,100),45,(0,0,255),cv2.FILLED)
plt.imshow(resim)
```

```python
cv2.putText(resim,"imdat",(0,250),cv2.FONT_HERSHEY_COMPLEX,2 ,(145,24,200))
plt.imshow(resim)
```

```python

# y ekseninde çevirme
yimg = cv2.flip(img, 1)
plt.imshow(yimg)
```

```python
#fiximg
adog=cv2.imread("/content/drive/MyDrive/sammy.jpg")
dog=cv2.cvtColor(adog,cv2.COLOR_BGR2RGB)
plt.imshow(dog)
```

```python
print(fiximg.shape)
print(dog.shape)
fiximg=cv2.resize(fiximg,(1025,1367))
print(fiximg.shape)
```

```python
#iki fotoğrafı birleştirme
birles=cv2.addWeighted(src1=fiximg,alpha=0.5,src2=dog,beta=0.5,gamma=0)
plt.imshow(birles)
```

```python
# ortalama ile bulanıklaştırm yöntemi
blurimg=cv2.blur(fiximg,ksize=(9,9))
plt.imshow(blurimg)
```

```python
# gaussian bulanıklaştırm yöntemi
gbBlur=cv2.GaussianBlur(fiximg,ksize=(9,9),sigmaX=7)
plt.imshow(gbBlur)
```

```python
# medyan ile bulanıklaştırm yöntemi
medyanblur=cv2.medianBlur(fiximg,ksize=15)
plt.imshow(medyanblur)
```

```python
#kenar belirleme -sadece x eksenlerini belirtme
sudoku=cv2.imread("/content/drive/MyDrive/sudoku.jpg")
sobelx=cv2.Sobel(sudoku,ddepth=cv2.CV_16S,dx=1,dy=0,ksize=5)
plt.imshow(sobelx)
```

```python
#kenarları belirtme
laplacian=cv2.Laplacian(sudoku,ddepth=cv2.CV_16S)
plt.imshow(laplacian)
```

```python
color=("b","g","r")
for i,c in enumerate(color):
  hist=cv2.calcHist([fiximg],channels=[i],mask= None,histSize=[256],ranges=[0,256])
  plt.plot(hist,color=c)
```
